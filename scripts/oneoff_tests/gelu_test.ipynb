```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Generate x values from -5 to 5
x = np.linspace(-5, 5, 1000)

# Compute the tanh, CDF, PDF, transformed sigmoid, GELU, and their derivatives for each x
y_tanh = np.tanh(x)
y_cdf = norm.cdf(x)
y_pdf = norm.pdf(x)
y_sigmoid_transformed = 2 / (1 + np.exp(-x)) - 1
y_gelu = 0.5 * x * (1 + np.tanh(np.sqrt(2 / np.pi) * (x + 0.044715 * x**3)))
y_sigmoid_transformed_derivative = 2 * np.exp(-x) / (1 + np.exp(-x))**2
y_gelu_derivative = 0.5 * (1 + np.tanh(np.sqrt(2 / np.pi) * (x + 0.044715 * x**3))) + 0.5 * x * (1 - np.tanh(np.sqrt(2 / np.pi) * (x + 0.044715 * x**3))**2) * np.sqrt(2 / np.pi) * (1 + 0.134145 * x**2)

# Create a single plot with all the functions and derivatives
plt.figure(figsize=(12, 8))
plt.plot(x, y_tanh, label='tanh')
plt.plot(x, y_cdf, label='CDF')
plt.plot(x, y_pdf, label='PDF')
plt.plot(x, y_sigmoid_transformed, label='Transformed Sigmoid')
plt.plot(x, y_gelu, label='GELU')
plt.plot(x, y_sigmoid_transformed_derivative, label='Transformed Sigmoid Derivative')
plt.plot(x, y_gelu_derivative, label='GELU Derivative')
plt.title('Activation Functions and Their Derivatives')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.ylim(-2, 2)
plt.show()
```
